```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('cleaned_medical_symptom_dataset1.csv')
print(df.head())
print("Missing values:\n", df.isnull().sum())
```

```python
empty_columns = df.columns[df.isnull().all()].tolist()
print("Empty columns:", empty_columns)

df_cleaned = df.drop(columns=empty_columns)

print("Cleaned DataFrame:")
print(df_cleaned.info())
```

```python
df_cleaned.isnull().sum() / len(df) * 100
df_cleaned.duplicated().sum()
```

```python
df_cleaned.drop_duplicates(inplace=True)
df_cleaned.duplicated().sum()
```

```python
df_cleaned.describe().T
```

```python
df_cleaned_objects = df_cleaned.select_dtypes(include=[object]).T
print(df_cleaned_objects)
```

```python
# Drop empty columns
df_cleaned.dropna(axis=1, how='all')

df_cleaned.columns
```

```python
columns_to_keep = ['text', 'label', 'symptoms', 'lebel_text']
df1 = df_cleaned.drop(columns=[col for col in df.columns if col not in columns_to_keep])
df1
```

```python
# Drop rows with missing values
df1 = df1.dropna()

# OR: Impute missing values
#df1.fillna(df1.mean(), inplace=True)  # Replace NaN with column mean for numeric data
df1
```

```python
# Standardize column names
df1.columns = df1.columns.str.strip().str.lower().str.replace(' ', '_')
df1 

```

```python
# Save to a new file
df1.to_csv('2422_rows_clean.csv', index=False)
```
